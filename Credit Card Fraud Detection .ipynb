```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_data = pd.read_csv('/content/creditcard.csv (1).zip')
credit_card_data
```

```python
#distribution of legit and fraudulent transaction
credit_card_data['Class'].value_counts()
```

```python
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
credit_card_data.groupby('Class').mean()
```

```python
#Under Sampling
legit_sample = legit.sample(n=492)
```

```python
new_set = pd.concat([legit_sample, fraud], axis=0)
new_set
```

```python
new_set.groupby('Class').mean()
```

```python
x=new_set.drop(columns='Class', axis=1)#features
y=new_set['Class']#targets
```

```python
x
```

```python
y
```

```python
#train test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
print(y.shape, y_train.shape, y_test.shape)
```

```python
#logistic regression
model=LogisticRegression()
```

```python
#train
model.fit(x_train, y_train)
```

```python
#Evaluate on accuracy
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
training_data_accuracy
```

```python
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
test_data_accuracy
```

```python
# test a random case

specific_case = new_set[new_set['Class'] == 1].iloc[0]
x_specific = specific_case.drop('Class')
y_specific = specific_case['Class']


prediction = model.predict([x_specific])


print("Predicted Class:", prediction[0])
print("Actual Class:", y_specific)
```
